```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from tqdm import tqdm
from PIL import Image
from io import BytesIO

# Categories and search terms
categories = {
    "nature": "beautiful nature scenery",
    "city": "modern city skyline"
}

# Create directories for images
for category in categories.keys():
    os.makedirs(category, exist_ok=True)

# Set up Selenium WebDriver
def setup_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Run without opening a browser
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1920x1080")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    return driver

# Function to download image
def download_image(url, folder, count):
    try:
        response = requests.get(url, timeout=5)
        img_data = response.content
        img_name = os.path.join(folder, f"{count}.jpg")
        
        # Save image
        with open(img_name, "wb") as f:
            f.write(img_data)
        return True
    except Exception as e:
        print(f"Failed to download {url}: {e}")
        return False

# Function to scrape images using Selenium
def scrape_images(search_term, folder, num_images=500):
    driver = setup_driver()
    search_url = f"https://www.google.com/search?q={search_term}&hl=en&tbm=isch"
    driver.get(search_url)
    time.sleep(2)

    # Scroll to load more images
    for _ in range(20):  # Scroll multiple times to load more images
        driver.find_element("tag name", "body").send_keys(Keys.PAGE_DOWN)
        time.sleep(1)

    # Parse page source with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    img_elements = soup.find_all("img")

    img_urls = [img['src'] for img in img_elements if 'src' in img.attrs]
    driver.quit()

    print(f"Found {len(img_urls)} images for {search_term}")

    # Download images
    count = 1
    for img_url in tqdm(img_urls[:num_images]):
        full_url = urljoin(search_url, img_url)
        if download_image(full_url, folder, count):
            count += 1
        if count > num_images:
            break

# Scrape images for both categories
for category, search_term in categories.items():
    scrape_images(search_term, category, num_images=500)

print("✅ Image scraping complete!")
```
